### Grouping by team

```python
# group by team and aggregate summary stats
df_team = df.groupby('team').agg(
    games_played=('ftr', 'count'),
    wins=('ftr', lambda x: (x == 'W').sum()),
    draws=('ftr', lambda x: (x == 'D').sum()),
    losses=('ftr', lambda x: (x == 'L').sum()),
    gf=('gf', 'sum'),
    ga=('ga', 'sum'),
).reset_index()

# add a goal differential column
df_team['gd'] = df_team.gf - df_team.ga

# add stats per game
df_team['win%'] = df_team.wins / df_team.games_played
df_team['draw%'] = df_team.draws / df_team.games_played
df_team['loss%'] = df_team.losses / df_team.games_played
df_team['avg_gf'] = df_team.gf / df_team.games_played
df_team['avg_ga'] = df_team.ga / df_team.games_played
df_team['avg_gd'] = df_team.gd / df_team.games_played

df_team
```

```python
sns.displot(data=df_team, x='games_played', kind='hist', kde=True, bins=10)
plt.title('Distribution of Games Played')
plt.show()
```

```python
sns.displot(data=df_team, x='wins', kind='hist', kde=True, bins=10)
plt.title('Distribution of Total Wins')
plt.show()
```

```python
sns.displot(data=df_team, x='win%', kind='hist', kde=True, bins=10)
plt.title('Distribution of Win Percentages')
plt.show()
```

**Findings**
* Most teams have fewer than 300 total games played in the league, but we see a handful of teams who have over 1000 games played
* Most teams have won less than 150 total games and win percentages have the highest concentration just under 30%
* There are far less draws than wins recorded, but the total number of draws are distributed more evenly with the highest concentation of draws being around 25% of games
* Losses are extremely high with most teams holding around a 45% loss record
* With the above points in mind, it makes sense that the average goal differential sits around -0.25 across the league
